# 2. bbox_valid 확인

### 필수 컬럼 무결성 점검

```python
# ==============================
# 1. 필수 컬럼 무결성 점검 (Must-Have)
# ==============================

essential_cols = [
    "image",
    "width",
    "height",
    "drug_N",
    "bbox_x",
    "bbox_y",
    "bbox_w",
    "bbox_h",
    "bbox_valid",
]

# 1-1. 결측치 개수 확인
print("결측치 개수:")
print(df_filtered[essential_cols].isnull().sum())

# 1-2. bbox 좌표가 없는 행 제거
before = len(df_filtered)
df_filtered = df_filtered.dropna(subset=["bbox_x", "bbox_y", "bbox_w", "bbox_h"])
after = len(df_filtered)
print(f"bbox 좌표 결측치 제거: {before} → {after}")

# 1-3. bbox_valid 필터링
print("bbox_valid 값 분포:\n", df_filtered["bbox_valid"].value_counts())
df_filtered = df_filtered[df_filtered["bbox_valid"] == True]
print("bbox_valid 필터링 후:", len(df_filtered))

# 1-4. 최종 shape
df_filtered.shape
```

# 3-1. 참고 메타데이터 분석

```python
# 3-1. dl_company (제약사별 분포)
plt.figure(figsize=(12, 4))
df_filtered["dl_company"].value_counts().head(20).plot(
    kind="bar", title="상위 20개 제약사"
)
plt.show()

# 3-2. form_code_name (제형 분포)
plt.figure(figsize=(12, 4))
df_filtered["form_code_name"].value_counts().head(20).plot(
    kind="bar", title="상위 20개 제형"
)
plt.show()

# 3-3. color_class1 (알약 색상 분포)
plt.figure(figsize=(12, 4))
df_filtered["color_class1"].value_counts().head(20).plot(
    kind="bar", title="상위 20개 색상"
)
plt.show()
```

# 3-2. 클래스 분포 확인 (drug_N 기준)

```python
# ==============================
# 3. 클래스 분포 확인 (drug_N 기준)
# ==============================
class_counts = df_filtered["drug_N"].value_counts()

plt.figure(figsize=(12, 5))
sns.histplot(class_counts, bins=50, kde=False)
plt.title("클래스별 라벨 개수 분포")
plt.xlabel("라벨 개수")
plt.ylabel("클래스 수")
plt.show()

print("총 클래스 수:", len(class_counts))
print("평균 샘플 수:", class_counts.mean())
print("최소 샘플 수:", class_counts.min())
print("최대 샘플 수:", class_counts.max())
```
