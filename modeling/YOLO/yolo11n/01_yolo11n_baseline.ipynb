# 이미지 1개 예측하기(테스트)

```python
results = model("./images/yolo_image.jpg", save=True)
result = results[0]
print(type(results))
print(len(results))
```

```python
result
```

```python
print(result.names)
print(result.boxes)
```

```python
boxes = result.boxes
boxes  # cls, conf, data
# cls 예측 클래스
# conf 예측 신뢰도
# data 박스좌표, 신뢰도, 클래스
# 길이 ===> 찾은 객체의 수
```

```python
print(type(boxes))
print(boxes.cls)  # 예측 결과 갯수 11개

print(
    boxes.conf
)  # 신뢰도 예측 신뢰도 11개 # tensor([0.9002] -> 90% 맞다) # 정확도로 이해하면 됨.

print(boxes.data)  # 박스좌표

boxes  # x1, y1, x2, y2, cls, conf
```

```python
# cls를 가져와서 반복문으로 names 의 value 찾아오기

names = results[0].names
print(names)
print(boxes.cls)

for idx in boxes.cls:  # 예측 결과 예시
    print(idx, int(idx))
    print(
        f"{names[int(idx)]}를 찾았습니다"
    )  # name에 어떤걸 찾을 수 있는 지 저장되어 있음
```
